```python
#Importar bibliotecas

import pandas as pd 
import plotly_express
```

```python
#Se carga el dataset de interés
data = pd.read_csv('vehicles_us.csv')
```

```python
data.info()
data.sample(5)
```

El dataset presenta información sobre vehículos de segunda mano dando información sobre el modelo del mismo, sus características físicas, la fecha en la que se publicó y el costo.
Se tienen 51,525 entradas y 12 columnas de las cuales hay valores ausentes en 5 columnas y en cada una el valor ausente se sustituirá con:
- Model_year: Se usará el 0 para sustituir los valores ausentes.
- cylinders: Se usará un 0 para sustituir los valores ausentes.
- odometer: Se usará el 0 para sustituir los valores ausentes.
- paint_color: Se usará 'Unknown' para sustituir los valores ausentes.
- is_4wd: Se tienen valores de 1 y NaN, por lo que se sustituirá NaN por 0.
        

```python
data['model_year'] = data['model_year'].fillna(0)
data['cylinders'] = data['cylinders'].fillna(0)
data['odometer'] = data['odometer'].fillna(0)
data['paint_color'] = data['paint_color'].fillna('Unknown')
data['is_4wd'] = data['is_4wd'].fillna(0)
```
